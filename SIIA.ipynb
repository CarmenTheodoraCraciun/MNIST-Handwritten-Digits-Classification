# Functia de evaluare a modelelor

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def evaluate_model(y_true, y_pred):
    '''
    FuncÈ›ia este conceputÄƒ pentru a evalua performanÈ›a unui model de clasificare prin calcularea diferitelor valori, cum ar fi acurateÈ›ea, raportul de clasificare È™i matricea de confuzie.

    Parametri:
       - y_true = Etichetele adevÄƒrate sau valorile de adevÄƒr de bazÄƒ.
       - y_pred = Etichetele estimate generate de model.
    '''
    # Calculeaza acuratetea
    accuracy = accuracy_score(y_true, y_pred)
    print("Accuracy:", accuracy)

    # Raportul de clasificare
    print("Classification report:")
    print(classification_report(y_true, y_pred))

    # Matricea de confuzie
    print("Confusion matrix:")
    print(confusion_matrix(y_true, y_pred))
```

* **AcurateÈ›ea** este raportul dintre numÄƒrul de predicÈ›ii corecte È™i numÄƒrul total de predicÈ›ii.
* **Raportul de clasificare** oferÄƒ o analizÄƒ detaliatÄƒ a performanÈ›ei modelului pentru fiecare clasÄƒ. Include metrici precum precizia, acoperirea (recall) È™i f1-score pentru fiecare clasÄƒ, precum È™i suportul (numÄƒrul de instanÈ›e din fiecare clasÄƒ).
  * *Precizia* - rata de predicÈ›ii corecte pentru o anumitÄƒ clasÄƒ.
  $precision = \frac{True Positives}{True Positives + False Positives}$
  * *Acoperire* - rata de identificare a instanÈ›elor relevante pentru o anumitÄƒ clasÄƒ.
  $recall = \frac{True Positives}{True Positives + False Negative}$
  * *F1-score* - o mÄƒsurÄƒ a performanÈ›ei unui model de clasificare È™i este definitÄƒ ca media armonicÄƒ Ã®ntre precizie È™i acoperire (recall). AceastÄƒ mÄƒsurÄƒ este utilÄƒ Ã®n special atunci cÃ¢nd existÄƒ un dezechilibru Ã®ntre clase Ã®n setul de date.
  $f1Score=2\frac{precision * recall}{precision + recall}$
  * *Numarul de instanÈ›e* din setul de date de testare care aparÈ›in unei clase.
  * *Macro Average* - media aritmeticÄƒ a preciziilor, acoperirii È™i F1-score-urile calculate pentru fiecare clasÄƒ individualÄƒ. Aceasta nu È›ine cont de dezechilibrul dintre clase È™i trateazÄƒ fiecare clasÄƒ Ã®n mod egal Ã®n calculul mediei.
  * *Weighted Average* - similar cu media macro, Ã®nsÄƒ fiecare clasÄƒ este ponderatÄƒ Ã®n funcÈ›ie de numÄƒrul de exemple din acea clasÄƒ Ã®n setul de date de testare. Acest lucru Ã®nseamnÄƒ cÄƒ clasele cu un numÄƒr mai mare de exemple vor avea o influenÈ›Äƒ mai mare asupra valorii medii.
* **Matricea de confuzie** este o matrice Ã®n care fiecare rÃ¢nd reprezintÄƒ clasa realÄƒ, iar fiecare coloanÄƒ reprezintÄƒ clasa prezisÄƒ. Elementul (i, j) al matricei este numÄƒrul de instanÈ›e ale clasei i care au fost clasificate ca aparÈ›inÃ¢nd clasei j. Diagonala principalÄƒ conÈ›ine numÄƒrul de clasificÄƒri corecte pentru fiecare clasÄƒ.

# Random Forest

Random Forest este un algoritm de Ã®nvÄƒÈ›are supervizatÄƒ care combinÄƒ mai mulÈ›i arbori de decizie pentru a obÈ›ine predicÈ›ii mai precise. FoloseÈ™te o tehnicÄƒ numitÄƒ "bagging" È™i construieÈ™te arbori de decizie pe subseturi aleatorii ale setului de date. Acest lucru Ã®l face robust È™i rezistent la overfitting. De asemenea, poate estima importanÈ›a fiecÄƒrei caracteristici din setul de date.

```python
from sklearn.ensemble import RandomForestClassifier

# Cream clasificatorul
classifier = RandomForestClassifier(n_estimators=100)

# Antrenam modelul
classifier.fit(X_train, y_train)

# Efectuarea de predicÈ›ii asupra datelor de testare
y_pred = classifier.predict(X_test)

evaluate_model(y_test, y_pred)
```

Din raportul de clasificare, pentru cifra:
* 0 avem o precizie de 97%, o acoperire de 99%, iar media armonicÄƒ dintre precizie È™i acoperire este de 98% din 980 de instanÈ›e.

# DNN - Dense Neural Networks

ReÈ›elele neuronale dense sunt o clasÄƒ de reÈ›ele neuronale artificiale Ã®n care neuroni (unitÄƒÈ›i de procesare) sunt aranjaÈ›i Ã®n straturi. Fiecare neuron dintr-un strat este conectat la toÈ›i neuroni din stratul anterior È™i la toÈ›i neuroni din stratul urmÄƒtor. Aceste conexiuni "dense" Ã®ntre straturi sunt ceea ce defineÈ™te reÈ›elele neuronale dense.

DNN-urile sunt capabile sÄƒ Ã®nveÈ›e caracteristici complexe din date, fÄƒcÃ¢ndu-le utile pentru o varietate de sarcini, cum ar fi recunoaÈ™terea imaginilor, clasificarea textului, regresia etc.

DeÈ™i DNN-urile sunt puternice, ele pot fi predispuse la overfitting (modelul Ã®nvaÈ›Äƒ sÄƒ identifice modele irelevante din datele de antrenament, ceea ce duce la o performanÈ›Äƒ slabÄƒ pe datele noi) pe seturile de date mici sau zgomotoase.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D

# Convertim cadrele de date in matrice
# Redimensionam datele pentru a se potrivi cu formularul de introducere a modelului (num_samples, 28, 28, 1)
X_train = X_train.values.reshape(-1, 28, 28, 1)
X_test = X_test.values.reshape(-1, 28, 28, 1)

# Normalizam datele
X_train, X_test = X_train / 255.0, X_test / 255.0
```

```python
# Cream modelul
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
# Compilam
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Antrenam
model.fit(X_train, y_train, epochs=5)
```

```python
y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
```

```python
evaluate_model(y_test, y_pred)
```

Modelul a prezis corect cifra din imaginile de testare Ã®n aproximativ 98.47% din cazuri.

Din raportul de clasificare, pentru cifra:
* 0 avem o precizie de 97%, o acoperire perfectÄƒ, iar media armonicÄƒ dintre precizie È™i acoperire este de 98% din 980 de instanÈ›e.
* 5 avem precizie perfectÄƒ, o acoperire de 98%, iar media armonicÄƒ dintre precizie È™i acoperire este de 99% din 892 de instanÈ›e.

Din matricea de confuzie, pentru cifra:

  * 976 de predicÈ›ii corecte, adicÄƒ cifra 0 a fost de 976 corect clasificatÄƒ (cazuri adevÄƒrat pozitive);
  * cÃ¢te o predicÈ›ie greÈ™itÄƒ ca 6, 7, 8 È™i 9, adicÄƒ cifra 0 a fost intuitÄƒ Ã®n cÃ¢te un caz ca cele patru cifre enumerate (cazuri fals negative);
